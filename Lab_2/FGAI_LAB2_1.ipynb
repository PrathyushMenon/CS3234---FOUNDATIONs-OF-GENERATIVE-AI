```python
import pandas as pd
df = pd.read_csv('/content/amazon.csv')
df.head()
```

```python
df.info()
print(f"DataFrame shape: {df.shape}")
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
print("Cleaned 'discounted_price' and 'actual_price' columns.")
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)
print("Cleaned 'discount_percentage' column.")
```

```python
print(df['rating'].unique())
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
print("Cleaned 'rating' column by converting to numeric and coercing errors to NaN.")
```

```python
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False).astype(float)
print("Cleaned 'rating_count' column.")
```

```python
print(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Imported matplotlib.pyplot as plt and seaborn as sns.")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['discount_percentage'], bins=20, kde=True)
plt.title('Distribution of Discount Percentage')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
print("Histogram of 'discount_percentage' displayed.")
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['discounted_price'])
plt.title('Box Plot of Discounted Price Distribution')
plt.ylabel('Discounted Price')
plt.show()
print("Box plot of 'discounted_price' displayed.")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['rating'], y=df['rating_count'])
plt.title('Scatter Plot of Rating vs. Rating Count')
plt.xlabel('Rating')
plt.ylabel('Rating Count')
plt.show()
print("Scatter plot of 'rating' versus 'rating_count' displayed.")
```

```python
category_counts = df['category'].astype(str).apply(lambda x: x.split('|')[0]).value_counts()
top_categories = category_counts.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Frequency of Top Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print("Bar chart of top product categories displayed.")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['actual_price'], bins=20, kde=True)
plt.title('Distribution of Actual Price')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()
print("Histogram of 'actual_price' displayed.")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
print("Histogram of 'rating' displayed.")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['actual_price'], y=df['discount_percentage'])
plt.title('Actual Price vs. Discount Percentage')
plt.xlabel('Actual Price')
plt.ylabel('Discount Percentage')
plt.show()
print("Scatter plot of 'actual_price' versus 'discount_percentage' displayed.")
```

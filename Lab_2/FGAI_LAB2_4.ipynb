```python
def generate_lol_text(prompt, max_new_tokens=100, num_return_sequences=1):
    """
    Generates text based on a given prompt using the loaded GPT-2 model.

    Args:
        prompt (str): The starting text for generation.
        max_new_tokens (int): The maximum number of new tokens to generate.
        num_return_sequences (int): The number of independent sequences to generate.

    Returns:
        list: A list of generated text strings.
    """
    generated_texts = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        num_return_sequences=num_return_sequences,
        truncation=True
    )
    return [g['generated_text'] for g in generated_texts]

print("Text generation function 'generate_lol_text' defined with max_new_tokens parameter.")
```

```python
lol_prompt = """A new champion joined the League: Zylos, the Chronomancer. His abilities include manipulating time to...
"""
generated_lol_story = generate_lol_text(lol_prompt, max_new_tokens=200, num_return_sequences=1)

print("Generated LoL Story (with max_new_tokens):")
for story in generated_lol_story:
    print(story)
```

```python
def generate_lol_text(prompt, max_new_tokens=100, num_return_sequences=1, temperature=0.7, top_k=50):
    """
    Generates text based on a given prompt using the loaded GPT-2 model.

    Args:
        prompt (str): The starting text for generation.
        max_new_tokens (int): The maximum number of new tokens to generate.
        num_return_sequences (int): The number of independent sequences to generate.
        temperature (float): The value used to modulate the next token probabilities. Higher values make output more random.
        top_k (int): The number of highest probability vocabulary tokens to keep for top-k-filtering.

    Returns:
        list: A list of generated text strings.
    """
    generated_texts = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        num_return_sequences=num_return_sequences,
        truncation=True,
        temperature=temperature,
        top_k=top_k
    )
    return [g['generated_text'] for g in generated_texts]

print("Text generation function 'generate_lol_text' updated with temperature and top_k parameters.")
```

```python
lol_prompt = """A new champion joined the League: Zylos, the Chronomancer. His abilities include manipulating time to...
"""
generated_lol_story = generate_lol_text(lol_prompt, max_new_tokens=200, num_return_sequences=1, temperature=0.8, top_k=60)

print("Generated LoL Story (with temperature and top_k):")
for story in generated_lol_story:
    print(story)
```

```python
lol_prompts = [
    "A new champion joined the League: Seraphina, the Echoing Song. Her abilities revolve around harmonizing with her allies to...",
    "A legendary item, the 'Hextech Flasher' was created in Zaun by a brilliant but eccentric inventor. It granted its wielder the ability to...",
    "The vast desert of Shurima holds ancient secrets, and deep beneath the sands, whispers tell of a forgotten ascended emperor who seeks to..."
]

print("League of Legends Prompts:")
for prompt in lol_prompts:
    print(f"- {prompt}")
```

```python
print("\n--- Generating text for each prompt ---")

for i, prompt in enumerate(lol_prompts):
    print(f"\nPrompt {i+1}: {prompt}")

    # Experiment with different parameters for each prompt
    if i == 0:
        # First prompt: balanced generation
        generated_story = generate_lol_text(prompt, max_new_tokens=180, temperature=0.8, top_k=70)
    elif i == 1:
        # Second prompt: more creative/random generation
        generated_story = generate_lol_text(prompt, max_new_tokens=220, temperature=0.9, top_k=80)
    else:
        # Third prompt: more focused/less random generation
        generated_story = generate_lol_text(prompt, max_new_tokens=200, temperature=0.7, top_k=60)

    print("Generated LoL Story:")
    for story in generated_story:
        print(story)
    print("\n--------------------------------------")
```
